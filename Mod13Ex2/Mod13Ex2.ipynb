```python
# 2
modelo = '''np.log(renda) ~ 
                               C(sexo, Treatment(0))
                               + C(educacao, Treatment(0))
                               + C(posse_de_imovel, Treatment(1))
                               + idade
                               + tempo_emprego
                               '''
md = smf.ols(modelo, data=test)
reg = md.fit_regularized(method='elastic_net', refit=True, L1_wt=0, alpha=0)

print(reg.params)
```
