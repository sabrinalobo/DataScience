# Tarefa 1

### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
sinasc_1 = sinasc_raw[['IDADEMAE', 'IDADEPAI', 'munResNome']]
sinasc_1.head()
```

```python
sinasc_1.groupby('munResNome').mean()
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
sinasc_2 = sinasc_raw[['PESO','DTNASC','ESCMAE']]
sinasc_2.head()
```

```python
sinasc_2.groupby(['DTNASC','ESCMAE']).mean()
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc_raw.loc[:,'DTNASC'] = pd.to_datetime(sinasc_raw['DTNASC'])
sinasc_3 = sinasc_raw[['IDADEMAE','IDADEPAI','munResNome','QTDFILVIVO']]
sinasc_3['ano'] = sinasc_raw['DTNASC'].apply(lambda i: i.year)
sinasc_3.head()
```

```python
def top(df, n = 5, column = 'QTDFILVIVO'):
    return df.fillna(0).sort_values(by = column, ascending = True)[:1]
```

```python
sinasc_3.groupby(['munResNome','ano'])['QTDFILVIVO'].count().sort_values(ascending=True)[:5]
```

O município 'Municipio ignorado - RO' foi o que nasceu menos, mas deve ser uma falta de valor, então vamos considerar 'Castanheiras'.

```python
sinasc_3[sinasc_3['munResNome'] == 'Castanheiras'].groupby('munResNome').aggregate(['mean', max, min])
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
sinasc_4 = sinasc_raw[['munResNome', 'QTDFILVIVO','IDADEPAI']]
sinasc_4['mês'] = sinasc_raw['DTNASC'].apply(lambda i: i.month)
sinasc_4.head()
```

```python
sinasc_4[sinasc_4['mês']==3].groupby('munResNome')['QTDFILVIVO'].count().sort_values(ascending=False)[:5]
```

Será Porto Velho

```python
sinasc_4[sinasc_4['munResNome'] == 'Porto Velho'].groupby('munResNome')['QTDFILVIVO','IDADEPAI'].aggregate(['mean', max, min])
```
